```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Opens and reads data.txt into the variable data
with open('data.txt', 'r') as file:
    data = np.array([[int(num) for num in line.strip().split(',')] for line in file.readlines()])
    
dleft  = data[:, 1] # distance of left motor
dright = data[:, 2] # distance of right motor
vleft  = dleft[1:]  - dleft[:-1] # velocity of left motor
vright = dright[1:] - dright[:-1] # velocity of right motor
u = data[1:, 0].reshape((-1, 1)) # input/duty cycle

# Plots the raw data
plt.plot(u, vleft, 'bo',  u, vright, 'yo')
plt.xlabel("U (PWM input)")
plt.ylabel("Velocity of Wheels")
plt.legend(("left", "right"), loc=0)
plt.title("Velocity of wheels vs. U (PWM input)")
plt.show()
```
